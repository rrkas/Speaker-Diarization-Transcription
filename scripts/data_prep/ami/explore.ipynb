```python
import os
import json
import numpy as np
from datasets import load_dataset

from scipy.io.wavfile import write
from pathlib import Path

# rootdir = (Path(__file__).parent.parent.parent.parent).resolve()
rootdir = Path("/home/rohnak.agarwal/projects/mtech-diarz")
print(rootdir)

cache_dir = (rootdir / "data" / "amicorpus").resolve()
print(cache_dir)
```

```python
os.makedirs(cache_dir, exist_ok=True)
ds = load_dataset("diarizers-community/ami", "ihm", cache_dir=str(cache_dir))
print(ds)
```

```python
ds["train"][0]
```

```python
def save_row(split: str, row: dict):
    fname = os.path.basename(row["audio"]["path"])
    audio_array = row["audio"]["array"]
    sampling_rate = row["audio"]["sampling_rate"]

    recs = [
        {
            "start": start,
            "end": end,
            "speaker": speaker,
        }
        for start, end, speaker in zip(
            row["timestamps_start"], row["timestamps_end"], row["speakers"]
        )
    ]

    audio_filepath = cache_dir / "simplified" / split / fname
    info_filepath = cache_dir / "simplified" / split / fname.replace(".wav", ".json")

    os.makedirs(audio_filepath.parent, exist_ok=True)
    write(audio_filepath, sampling_rate, audio_array.astype(np.float32))

    with open(info_filepath, "w") as f:
        json.dump(recs, f, indent=4)
```

```python
save_row(ds["train"][0])
```
